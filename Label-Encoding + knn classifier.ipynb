```python
class CFG:
    SEED = 123
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(CFG.SEED) # Seed 고정
```

## Data Load

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
for idx in range(1,10):
    for i in range(len(train)):
        train.loc[i,f'SNP_0{idx}_x'] = train.loc[i,f'SNP_0{idx}'].split()[0]
        train.loc[i,f'SNP_0{idx}_y'] = train.loc[i,f'SNP_0{idx}'].split()[1]

for idx in range(10,16):
    for i in range(len(train)):
        train.loc[i,f'SNP_{idx}_x'] = train.loc[i,f'SNP_{idx}'].split()[0]
        train.loc[i,f'SNP_{idx}_y'] = train.loc[i,f'SNP_{idx}'].split()[1]


for idx in range(1,10):
    for i in range(len(test)):
        test.loc[i,f'SNP_0{idx}_x'] = test.loc[i,f'SNP_0{idx}'].split()[0]
        test.loc[i,f'SNP_0{idx}_y'] = test.loc[i,f'SNP_0{idx}'].split()[1]

for idx in range(10,16):
    for i in range(len(test)):
        test.loc[i,f'SNP_{idx}_x'] = test.loc[i,f'SNP_{idx}'].split()[0]
        test.loc[i,f'SNP_{idx}_y'] = test.loc[i,f'SNP_{idx}'].split()[1]
```

```python
le = preprocessing.LabelEncoder()
tar = ['SNP_01', 'SNP_02',
       'SNP_03', 'SNP_04', 'SNP_05', 'SNP_06', 'SNP_07', 'SNP_08', 'SNP_09',
       'SNP_10', 'SNP_11', 'SNP_12', 'SNP_13', 'SNP_14', 'SNP_15',
      'SNP_01_x', 'SNP_01_y', 'SNP_02_x', 'SNP_02_y', 'SNP_03_x',
       'SNP_03_y', 'SNP_04_x', 'SNP_04_y', 'SNP_05_x', 'SNP_05_y',
       'SNP_06_x', 'SNP_06_y', 'SNP_07_x',
       'SNP_07_y',  'SNP_08_x', 'SNP_08_y', 'SNP_09_x', 'SNP_09_y', 
       'SNP_10_x', 'SNP_10_y',  'SNP_11_x', 'SNP_11_y', 'SNP_12_x', 'SNP_12_y',
       'SNP_13_x', 'SNP_13_y', 'SNP_14_x', 'SNP_14_y', 'SNP_15_x', 'SNP_15_y']
for col in tar:
    le.fit(test[col])
    test[col] = le.transform(test[col])
    
tar = ['SNP_01', 'SNP_02',
       'SNP_03', 'SNP_04', 'SNP_05', 'SNP_06', 'SNP_07', 'SNP_08', 'SNP_09',
       'SNP_10', 'SNP_11', 'SNP_12', 'SNP_13', 'SNP_14', 'SNP_15',
      'SNP_01_x', 'SNP_01_y', 'SNP_02_x', 'SNP_02_y', 'SNP_03_x',
       'SNP_03_y', 'SNP_04_x', 'SNP_04_y', 'SNP_05_x', 'SNP_05_y',
       'SNP_06_x', 'SNP_06_y', 'SNP_07_x',
       'SNP_07_y',  'SNP_08_x', 'SNP_08_y', 'SNP_09_x', 'SNP_09_y', 
       'SNP_10_x', 'SNP_10_y',  'SNP_11_x', 'SNP_11_y', 'SNP_12_x', 'SNP_12_y',
       'SNP_13_x', 'SNP_13_y', 'SNP_14_x', 'SNP_14_y', 'SNP_15_x', 'SNP_15_y','class']
for col in tar:
    le.fit(train[col])
    train[col] = le.transform(train[col])
```

```python
def get_x_y(df):
    if 'class' in df.columns:
        df_x = df.drop(columns=['id', 'class'])
        df_y = df['class']
        return df_x, df_y
    else:
        df_x = df.drop(columns=['id'])
        return df_x
```

```python
train_x, train_y = get_x_y(train)
test_x = get_x_y(test)
```

## Data Pre-processing
### Label-Encoding

```python
train_x
```

```python
sc = StandardScaler()
sc.fit(train_x)
X_train_std = sc.transform(train_x)
X_test_std = sc.transform(test_x)
```

```python
mm = MinMaxScaler()
mm.fit(train_x)
X_train_std = mm.transform(train_x)
X_test_std = mm.transform(test_x)
```

## Model Fit

```python
Kn = KNeighborsClassifier(n_neighbors = 5, p = 2, metric='minkowski', weights='distance')
Kn.fit(X_train_std, train_y)
```

```python
clf = RandomForestClassifier(random_state=CFG.SEED, n_estimators=2000)
clf.fit(train_x, train_y)
```

## Inference

```python
preds = Kn.predict(X_test_std)
print('KN Done.')
```

```python
preds = clf.predict(test_x)
print('CLF Done.')
```
